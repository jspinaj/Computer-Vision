```python
def umb_adap(imagen,sz,c):
    imagen=cv2.cvtColor(imagen,cv2.COLOR_RGB2YCrCb)
    Y,Cr,Cb = cv2.split(imagen)
    th1 = cv2.adaptiveThreshold(Cr,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C,cv2.THRESH_BINARY,sz,c)
    ima= cv2.bitwise_and(Cr,th1)
```
